# Data description

**events_train.csv** -information about students steps activity

1. **step_id** - id of step  
2. **user_id** - id of user  
3. **timestamp** - time of the event in unix date  
4. **action** - event, possible values:  
*discovered* - user went to a step  
*viewed* - step preview  
*started_attempt* - started attempt to pass step   
*passed* - Correct solution to a practical step  
  
**submissions_train.csv** - information about time and status of submits to practical steps

1. **step_id** - id of step  
2. **timestamp** - time of sending submit in unix date  
3. **submission_status** - status of submission  
4. **user_id** - id of user

## events_train.csv

```python
events_data['date'] = pd.to_datetime(events_data['timestamp'], unit='s')
events_data['day'] = events_data['date'].dt.date
events_data.head()
```

-**For each step we will count its 'difficulty' by 'passed, discovered, started_attempt, viewed' statistics**

```python
event_steps = events_data.pivot_table(index='step_id', columns='action', values='user_id',
                       aggfunc='count', fill_value=0).reset_index()
event_steps['passed / discovered'] = event_steps['passed'] / event_steps['discovered']
event_steps['passed / started_attempt'] = event_steps['passed'] / event_steps['started_attempt']
event_steps['passed / viewed'] = event_steps['passed'] / event_steps['viewed']
event_steps = event_steps.replace([np.inf, -np.inf], np.nan)
event_steps = event_steps.fillna(1)
event_steps
```

```python
events_data
```

```python
events_data = events_data.merge(event_steps[['step_id',
                               'passed / discovered',
                               'passed / started_attempt', 'passed / viewed']], on='step_id', how='outer')
events_data
```

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, classification_report
from sklearn.model_selection import cross_val_score
```

```python
poly = PolynomialFeatures(degree=3)
X = poly.fit_transform(X)
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
y.value_counts()
```

```python
cols = ['model', 'train_accuracy_score', 'train_roc_auc_score', 'train_mean_cross_val_score', 'train_proba_roc_auc_score',
       'test_accuracy_score', 'test_roc_auc_score','test_proba_roc_auc_score', 'test_mean_cross_val_score']
model_performance = pd.DataFrame(columns=cols)
```

```python
def add_performance_to_df(df_name, name_model, model, train_X, train_y, test_X, test_y):
    adder = {'model' : '', 'train_accuracy_score': '', 'train_roc_auc_score': '', 'train_proba_roc_auc_score': '',
             'train_mean_cross_val_score': '', 'test_accuracy_score': '',
             'test_roc_auc_score': '','test_proba_roc_auc_score': '', 'test_mean_cross_val_score': ''}
    
    train_predictions = model.predict(train_X)
    train_proba_predictions = model.predict_proba(train_X)[:, 1]
    test_predictions = model.predict(test_X)
    test_proba_predictions = model.predict_proba(test_X)[:, 1]
    
    adder['model'] = name_model
    adder['train_accuracy_score'] = accuracy_score(train_y, train_predictions)
    adder['test_accuracy_score'] = accuracy_score(test_y, test_predictions)
    adder['train_roc_auc_score'] = roc_auc_score(train_y, train_predictions)
    adder['train_proba_roc_auc_score'] = roc_auc_score(train_y, train_proba_predictions)
    adder['test_roc_auc_score'] = roc_auc_score(test_y, test_predictions)
    adder['test_proba_roc_auc_score'] = roc_auc_score(test_y, test_proba_predictions)
    adder['train_mean_cross_val_score'] = cross_val_score(model, train_X, train_y, cv=5).mean()
    adder['test_mean_cross_val_score'] = cross_val_score(model, test_X, test_y, cv=5).mean()
    
    
    return df_name.append(adder, ignore_index=True)
    
```

## DecisionTreeClassifier

#### train test split

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(random_state=17, class_weight='balanced')
X.shape, y.shape
```

```python
dtc.fit(X_train_scaled, y_train)
```

```python
model_performance = add_performance_to_df(model_performance, 'DecisionTreeClassifier_train_test',
                      dtc, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
test_clas_rep_dtc_train_test = pd.DataFrame(classification_report(y_test, dtc.predict(X_test_scaled), output_dict=True))
```

#### GridSearchCV

```python
skf = StratifiedKFold(shuffle=True, random_state=17)

dtc_params = {'criterion': ['entropy', 'gini'],
         'max_depth': range(1, 10),
         'min_samples_split': range(2, 8, 2),
         'min_samples_leaf': range(1, 5)}

dtc_gs = GridSearchCV(estimator=DecisionTreeClassifier(random_state=17, class_weight='balanced'), param_grid=dtc_params,
                      scoring='roc_auc',
                cv=skf, n_jobs=-1, verbose=True)
dtc_gs.fit(X_train_scaled, y_train)
```

```python
best_dtc = dtc_gs.best_estimator_
model_performance = add_performance_to_df(model_performance, 'DecisionTreeClassifier_GridSearchCV',
                      best_dtc, X_train_scaled, y_train, X_test_scaled, y_test)
test_clas_rep_dtc_cv = pd.DataFrame(classification_report(y_test, best_dtc.predict(X_test_scaled), output_dict=True))
```
